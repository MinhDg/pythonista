```python
# How to fill in for missing values
```

```python
np.random.seed(123)
df_obj = DataFrame(np.random.rand(36).reshape(6,6))
df_obj
```

```python
# select rows and comlumn 
# assign them into missing values

df_obj.loc[3:5, 0] = missing
df_obj.loc[1:4, 5]= missing
df_obj
```

```python
# use the fillna method to fill our NaN values with a value of zero
fill_df = df_obj.fillna(0)
fill_df
```

```python
# Into the fillna method. 
# This will then fill in the missing values from each column series as designated by the dictionary key with its own value as specified by the corresponding dictionary value.
# create a dictionary and then for the values in column at index position zero,
    # we're going to set those equal to 0.1. 
    # for the missing values in the column at index position five,  set those equal to 1.25. 




fill_df = df_obj.fillna({0 : 0.1, 5 : 1.25})
fill_df
```

```python
# the missing values have now been filled in column, at index position zero, 
# the missing values have been filled with .1 and then at index position five, 
# they've been filled in with 1.25. 
```

```python
# COUNTING missing values

# count missing values in order to figure out which variables are most problematic.
   # In other words, which variables in your dataset have the most number of missing values
```

```python
np.random.seed(123)
df_obj = DataFrame(np.random.rand(36).reshape(6,6))
df_obj.loc[3:5, 0] = missing
df_obj.loc[1:4, 5]= missing
df_obj
```

```python
# count missing values per column

df_obj.isnull().sum()
```

```python
# in the column at index position zero, we have three null values 
# at index position five we have four null values 
# in between, we have no null values in any of the other columns
```

```python
# Filtering out missing values
df_no_Na = df_obj.dropna()
df_no_Na 
```

```python
#  all of the rows that had a null value, any null value, were all dropped 
# and so all we have left now is just one row, that's from index position zero
```

```python
# if you wanted to drop the columns that contain any missing values instead of dropping the row

df_no_NaN = df_obj.dropna(axis = 1)
df_no_NaN
```
